```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm, uniform, expon, binom, poisson, geom, describe, mode
```

# Генерация распределений

```python
data_norm = pd.Series(norm.rvs(size=1000))
data_uniform = pd.Series(uniform.rvs(size=1000))
data_expon = pd.Series(expon.rvs(size=1000))
data_binom = pd.Series(binom.rvs(n=10, p=0.3, size=1000))
data_poisson = pd.Series(poisson.rvs(mu=3, size=1000))
data_geom = pd.Series(geom.rvs(p=0.3, size=1000))
```

```python
assert len(data_norm) == len(data_uniform) == len(data_expon) == len(data_binom) == len(data_poisson) \
       == len(data_geom)
```

```python
data = pd.concat([data_norm, data_uniform, data_expon, data_binom, data_poisson, data_geom], 
                 axis=1).rename(columns={0: 'Нормальное',
                                         1: 'Равномерное',
                                         2: 'Экспоненциальное',
                                         3: 'Биномиальное',
                                         4: 'Пуассона',
                                         5: 'Геометрическое'})
```

# Одномерный анализ каждого столбца в виде гистограмм и статистических сводок

```python
data.hist(bins=20, figsize=(14,10));
```

```python
for column in data.columns:
    print(f'Статистическая сводка для столбца: {column}')
    print(f'Минимум и максимум: {describe(data[column])[1]}')
    print(f'Среднее: {describe(data[column])[2]}')
    print(f'Дисперсия: {describe(data[column])[3]}')
    print(f'Стандартное отклонение: {np.std(data[column])}')
    print(f'Мода: {mode(data[column], keepdims=True)[0][0]}')
    print('----------------------')
```
